```python
# Drop null values
# replace null values
```

```python
# object---> mode
# numerical--> mean/median
```

```python
col_list=df.columns
col_list
```

```python
# df["age"].mode()
df["age"].mean()
```

```python
for col in col_list:
    if df[col].dtype=="object":
        df[col]=df[col].fillna(df[col].mode()[0])
    else:
        df[col]=df[col].fillna(df[col].mean())
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Outliers
```

```python
for col in col_list:
    if df[col].dtype!="object":
        sns.boxplot(df[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
# Removal Of outliers
```

```python
df["pdays"].value_counts()
```

```python
df["previous"].value_counts()
```

```python
out_cols=["age","duration","campaign","cons.conf.idx"]
```

```python
for col in out_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    df=df[(df[col]>=lower)&(df[col]<=upper)]
```

```python
for col in col_list:
    if df[col].dtype!="object":
        sns.boxplot(df[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```
